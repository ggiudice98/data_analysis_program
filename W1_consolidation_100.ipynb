# SECTION 4 — Subqueries (Q66–Q100)

**Q66 [eco]** — Completed orders above the average completed order value. Subquery in WHERE.

```python
q66 = """   SELECT COUNT(order_id) AS completed_orders
            FROM orders 
            WHERE status = 'completed' AND total_amount > ( SELECT AVG(total_amount) AS avg_amount
                                                    FROM orders
                                                      WHERE status = 'completed'
                                              )
        

"""
df = pd.read_sql_query(q66, eco)
display(df)
```

**Q67 [eco]** — Customers whose total completed spend is above the average customer spend. Show name, segment, total_spend.

**Q100 — CAPSTONE (two parts)**

**Part A [eco]** — Full ecommerce segment report: segment, customer_count (COUNT DISTINCT), completed_orders (COUNT DISTINCT), item_revenue (SUM oi.quantity * oi.unit_price), avg_order_value at item grain, top_category per segment (correlated subquery), Above/Below company avg flag. Order by item_revenue desc.

**Part B [log]** — Full logistics driver report: driver name, province, total_shipments, total_freight, avg_freight per shipment (rounded 2dp), top_route (most freight, correlated subquery), Above/Below company avg freight flag. Delivered only. Order by total_freight desc.

```python
q100a = """
        WITH per_segment AS ( SELECT c.segment,
                                        COUNT(DISTINCT c.customer_id) AS customer_count,
                                         COUNT(DISTINCT o.order_id) AS order_count,
                                        SUM(oi.quantity * oi.unit_price) AS revenue,
                                        AVG(oi.quantity * oi.unit_price) AS avg_order_value
    FROM orders o JOIN customers c ON o.customer_id = c.customer_id JOIN order_items oi ON o.order_id = oi.order_id
    WHERE o.status = 'completed'
    GROUP BY c.segment),

    category_revenue AS ( SELECT c.segment,
                                oi.category,
                                SUM(oi.quantity * oi.unit_price) AS cat_revenue
    FROM orders o JOIN customers c ON o.customer_id = c.customer_id JOIN order_items oi ON o.order_id = oi.order_id
        WHERE o.status = 'completed'
     GROUP BY c.segment, oi.category),
   
   
   category_rank AS (
    SELECT segment,
           category,
           RANK() OVER (PARTITION BY segment ORDER BY cat_revenue DESC) AS rnk
    FROM category_revenue),

   avg AS (
    SELECT AVG(oi.quantity * oi.unit_price) AS company_avg
    FROM order_items oi JOIN orders o ON oi.order_id = o.order_id
     WHERE o.status = 'completed')
   
SELECT ps.segment,
       ps.customer_count,
       ps.order_count,
       ps.revenue,
       ROUND(ps.avg_order_value, 2) AS avg_order_value,
       cr.category AS top_category,
       (CASE WHEN ps.revenue > avg.company_avg THEN 'Above' ELSE 'Below' END) AS flag
FROM per_segment ps JOIN category_rank cr ON ps.segment = cr.segment AND cr.rnk = 1 CROSS JOIN avg
ORDER BY ps.revenue DESC

"""
df = pd.read_sql_query(q100a, eco)
display(df)
```

**Part B [log]** — Full logistics driver report: driver name, province, total_shipments, total_freight, avg_freight per shipment (rounded 2dp), top_route (most freight, correlated subquery), Above/Below company avg freight flag. Delivered only. Order by total_freight desc.

```python
# Part B — Logistics driver report
q100b = """
WITH per_driver AS (    SELECT  s.driver_id, 
                                d.name, 
                                d.province,
                                COUNT(shipment_id) AS total_shipments,
                                SUM(s.freight_charge) AS total_freight,
                                ROUND(AVG(freight_charge), 2) AS avg_freight
                         FROM shipments s JOIN drivers d ON s.driver_id = d.driver_id 
                        GROUP BY s.driver_id, d.name, d.province),

        per_route AS (  SELECT route_id,
                                driver_id,
                                SUM(freight_charge) AS  route_freight

                        FROM shipments
                        GROUP BY route_id, driver_id),
                 
        route_rank AS (SELECT   driver_id,
                                route_id,
                                route_freight,
                                RANK() OVER (PARTITION BY driver_id ORDER BY route_freight DESC) AS rnk
                                FROM per_route),
          
        avg AS ( SELECT avg(total_freight) AS company_avg
        FROM per_driver)             

        SELECT pd.name,
                pd.province,
                pd.total_shipments,
                pd.total_freight,
                pd.avg_freight,
                (CASE WHEN pd.total_freight > a.company_avg THEN 'Above' ELSE 'Below' END) AS flag
        FROM per_driver pd JOIN route_rank rr ON pd.driver_id = rr.driver_id CROSS JOIN avg a WHERE rr.rnk = 1
        ORDER BY total_freight DESC
                                
                       

                            
                         

"""
df = pd.read_sql_query(q100b, log)
display(df)
```
